# **Neural Network for Digit Classisfication**

```python
import tensorflow # Tensorflow is a open-source AI/ML Library developed by GOOGLE.
from tensorflow import keras # Keras is a high-level API inside Tensorflow that consist of pre-build tools and DL models.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Dataset
(X_train, Y_train), (X_test, Y_test) = keras.datasets.mnist.load_data()
```

```python
plt.matshow(X_train[1])
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train_flat = X_train.reshape(60000,28*28) #Reshape to make 3D to 2D
X_test_flat = X_test.reshape(10000,28*28)
```

```python
X_train_flat.shape
```

```python
X_test_flat.shape
```

```python
#Model Building
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='softmax') #One Layer
])
```

```python
#Model Compilation
model.compile(
    optimizer='adam', #ADAM is a optimizer like GRADIENT DESCENT
    loss='sparse_categorical_crossentropy', #It is a multi-class classification so we use CATEGORICAL CROSS ENTROPY.
    metrics=['accuracy']
)
```

```python
#Model Training
model.fit(X_train_flat, Y_train, epochs=5, batch_size=5000)
```

```python
model.evaluate(X_test_flat, Y_test)
```
